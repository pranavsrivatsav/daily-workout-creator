### Get workout dates, and the exercises done on that date and put it into a dictionary

```python
# get all the unique start times and strip the time from the date
# comma is the delimiter, get only the first part of the date
workout_data['start_time'] = workout_data['start_time'].str.split(',').str[0]
unique_start_times = workout_data['start_time'].unique()

display(unique_start_times)
```

```python
#for each of the unique start times, get the unique exercises done at that time
#put the unique exercises in a dictionary with the start time as the key and unique exercises in a list as the value
unique_exercises = {}
for start_time in unique_start_times:
    unique_exercises[start_time] = workout_data[workout_data['start_time'] == start_time]['exercise_title'].unique()

# display(unique_exercises)

#convert each value from array to list
for key in unique_exercises:
    unique_exercises[key] = list(unique_exercises[key])

#reverse the order of the dictionary
unique_exercises = dict(reversed(unique_exercises.items()))

display(unique_exercises)
```
